```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
df
```

```python
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='+')
```

```python
print(df.columns)
```

```python
# Extract feature and target
X = df[['year']]  # Independent variable
y = df['per capita income (US$)']  # Dependent variable
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Train the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predict for the year 2020
year_2020 = np.array([[2020]])
predicted_income = model.predict(year_2020)

print(f"Predicted Per Capita Income for Canada in 2020: ${predicted_income[0]:,.2f}")
```

```python
# Plot the regression line
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter([2020], predicted_income, color='green', marker='o', s=100, label='Prediction (2020)')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Canada Per Capita Income Prediction')
plt.legend()
plt.show()
```

```python
year_input = int(input("Enter the year to predict per capita income: "))
year_array = np.array([[year_input]])  # Convert input to a 2D array
predicted_income = model.predict(year_array)

print(f"Predicted Per Capita Income for Canada in {year_input}: ${predicted_income[0]:,.2f}")
```
